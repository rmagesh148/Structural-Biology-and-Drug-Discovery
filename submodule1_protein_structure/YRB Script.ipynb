```python
from pymol import cmd

def yrb(selection='all'):
    cmd.remove("hydro")
    cmd.set_color('yellow', [0.950, 0.78, 0.0])
    cmd.set_color('grey', [0.95, 0.95, 0.95])
    cmd.set_color('red', [1.0, 0.4, 0.4])
    cmd.set_color('blue', [0.2, 0.5, 0.8])

    mapping = {
        'arg': [('NE,NH2,NH1', 'blue'), ('CD,CZ', 'grey'), ('CG', 'yellow')],
        'asn': [('CG,OD1,ND2', 'grey')],
        'asp': [('CG', 'grey'), ('OD2,OD1', 'red')],
        'cys': [('SG', 'grey')],
        'gln': [('CG', 'yellow'), ('CD,OE1,NE2', 'grey')],
        'glu': [('CG', 'yellow'), ('CD', 'grey'), ('OE1,OE2', 'red')],
        'his': [('CG,CD2,ND1,NE2,CE1', 'grey')],
        'ile': [('CG1,CG2,CD1', 'yellow')],
        'leu': [('CG,CD1,CD2', 'yellow')],
        'lys': [('CG,CD', 'yellow'), ('CE', 'grey'), ('NZ', 'blue')],
        'met': [('CG,CE', 'yellow'), ('SD', 'grey')],
        'phe': [('CG,CD1,CE1,CZ,CE2,CD2', 'yellow')],
        'pro': [('CG', 'yellow'), ('CD', 'grey')],
        'ser': [('CB,OG', 'grey')],
        'thr': [('CB,OG1', 'grey'), ('CG2', 'yellow')],
        'trp': [('CG,CD2,CZ2,CH2,CZ3,CE3', 'yellow'), ('CD1,NE1,CE2', 'grey')],
        'tyr': [('CG,CE1,CD1,CE2,CD2', 'yellow'), ('CZ,OH', 'grey')],
        'val': [('CG1,CG2', 'yellow')]
    }

    obj_list = cmd.get_names('objects')
    for obj in obj_list:
        if (obj == selection or selection == 'all'):
            cmd.color('grey', '(n. N,C,CA,O and ' + obj + ')')
            cmd.color('yellow', '(n. CB and ' + obj + ')')

            for key in mapping:
                for (atom, color) in mapping[key]:
                    cmd.color(color, '( n. ' + atom + ' and r. ' + key + ' and ' + obj + ' )')

cmd.extend('yrb', yrb)
```
