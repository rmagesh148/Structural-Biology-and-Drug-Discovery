In this tutorial, we will explore how to apply colors to different parts of a protein structure using PyMOL to highlight specific features. Weâ€™ll cover:

Secondary Structure Coloring: Color different parts of the secondary structure (Î±-helices, Î²-sheets, loops).
Chain-Specific Coloring: Assign colors to distinguish different chains.
Element-Based Coloring: Color atoms by element type or chemical properties.
Let's get started with a concise step-by-step guide for each of these tasks.

## **I. GUI Steps**


### **A. Secondary Structure Coloring:**

1. Click on the C (Color) button next to the object in the right panel.
2. Select By SS (Secondary Structure) > Select the kind of color you want for Helix,Sheet and Loop.

This will automatically assign colors to Î±-helices, Î²-sheets, and loops.

### **B. Chain-Specific Coloring:**

1. Click on the C (Color) button next to the object.
2. Select By Chain from the dropdown menu.
3. Choose between different options:
- By Chain (elem C) or By Chain (*/CA) to color based on different chain elements or chain identifiers.
- You can also use Chainbows if you want a rainbow-like effect across different chains.

You can pick specific colors for each chain, such as red for Chain A and blue for Chain B, from the color panel below.

### **C. Element-Based Coloring:**
1. Click the C (Color) button next to the object.
2. Select By Element from the dropdown menu.
3. This will assign colors based on the element type:
- Carbon (C): Generally gray by default.
- Oxygen (O): Red.
- Nitrogen (N): Blue.
- Sulfur (S): Yellow.

You can also manually change the color by selecting the desired element type and picking a color from the color options.


## **II.Command Line Steps**

Using the command line allows for precise and rapid modifications. Here are the steps to apply different coloring schemes:

```python
# Secondary Structure Coloring
color yellow, ss h  # Color Î±-helices yellow
color cyan, ss s    # Color Î²-sheets cyan
color green, ss l   # Color loops green
```

```python
# Chain-Specific Coloring
color red, chain A  # Color Chain A red
color blue, chain B  # Color Chain B blue
```

```python
# Element-Based Coloring
color red, elem O   # Color oxygen atoms red
color blue, elem N  # Color nitrogen atoms blue
color gray, elem C  # Color carbon atoms gray
```

```python
import os
from Bio.PDB import PDBParser, NeighborSearch, Selection
from Bio.PDB.PDBExceptions import PDBConstructionWarning
from Bio.SeqUtils import seq1
import nglview as nv
import warnings
import requests
```

```python
# Function to download PDB file
def download_pdb_file(pdb_id):
    url = f'https://files.rcsb.org/download/{pdb_id}.pdb'
    response = requests.get(url)
    file_path = f'{pdb_id}.pdb'
    with open(file_path, 'w') as file:
        file.write(response.text)
    return file_path

# Function to parse PDB file and extract structure
def parse_pdb_file(file_path):
    parser = PDBParser(QUIET=True)
    with warnings.catch_warnings():
        warnings.simplefilter('ignore', PDBConstructionWarning)
        structure = parser.get_structure('protein', file_path)
    return structure

# Function to analyze protein-ligand interactions
def analyze_protein_ligand_interactions(structure, ligand_resname):
    model = structure[0]
    # Check for available chains
    chains = list(model.get_chains())
    chain = None
    for ch in chains:
        if ligand_resname in [res.get_resname() for res in ch.get_residues()]:
            chain = ch
            break
    if not chain:
        raise ValueError(f"Ligand with residue name {ligand_resname} not found in any chain.")

    atoms = Selection.unfold_entities(model, 'A')
    ns = NeighborSearch(atoms)

    ligand_atoms = [atom for atom in chain.get_atoms() if atom.get_parent().get_resname() == ligand_resname]
    interactions = {}

    for ligand_atom in ligand_atoms:
        neighbors = ns.search(ligand_atom.coord, 4.0)  # 4.0 Ã… cutoff for interaction
        for neighbor in neighbors:
            res_id = neighbor.get_parent().get_id()
            if res_id[0] == ' ':  # Only consider standard residues
                aa_name = seq1(neighbor.get_parent().get_resname())
                aa_position = neighbor.get_parent().get_id()[1]
                interaction_type = classify_interaction(ligand_atom, neighbor)
                if res_id not in interactions:
                    interactions[res_id] = []
                interactions[res_id].append((ligand_atom, neighbor, aa_name, aa_position, interaction_type))

    print(f"Found {len(interactions)} interacting residues.")
    for res_id, atom_pairs in interactions.items():
        for ligand_atom, protein_atom, aa_name, aa_position, interaction_type in atom_pairs:
            print(f"Ligand atom {ligand_atom.get_name()} interacts with {aa_name} at position {aa_position}, atom {protein_atom.get_name()} with interaction type: {interaction_type}")

    return interactions

# Function to classify the type of interaction based on distance and atom types
def classify_interaction(ligand_atom, protein_atom):
    distance = ligand_atom - protein_atom
    if distance < 2.5:
        return "Covalent"
    elif distance < 3.5:
        return "Hydrogen bond"
    else:
        return "Van der Waals"

# Function to visualize protein-ligand interactions using NGLView
def visualize_protein_ligand_interactions(pdb_file, interactions):
    view = nv.show_file(pdb_file)
    view.add_representation('cartoon', selection='protein', color='blue')
    view.add_representation('ball+stick', selection='ligand', color='red')

    for res_id, atom_pairs in interactions.items():
        for ligand_atom, protein_atom, aa_name, aa_position, interaction_type in atom_pairs:
            view.add_distance({
                'atom1': {
                    'chain': ligand_atom.get_parent().get_parent().get_id(),
                    'residue_number': ligand_atom.get_parent().get_id()[1],
                    'atom': ligand_atom.get_name()
                },
                'atom2': {
                    'chain': protein_atom.get_parent().get_parent().get_id(),
                    'residue_number': protein_atom.get_parent().get_id()[1],
                    'atom': protein_atom.get_name()
                },
                'color': 'yellow'
            })

    view.center()
    return view

# Main function
def main():
    pdb_id = "1a3n"  # Example PDB ID for hemoglobin
    ligand_resname = "HEM"  # Example ligand residue name for hemoglobin

    # Download PDB file
    pdb_file = download_pdb_file(pdb_id)

    # Parse PDB file
    structure = parse_pdb_file(pdb_file)

    # Analyze protein-ligand interactions
    interactions = analyze_protein_ligand_interactions(structure, ligand_resname)

    # Visualize the interactions
    view = visualize_protein_ligand_interactions(pdb_file, interactions)
    return view

# Execute the main function and display the view
view = main()
view
```

## **II. Command Line Steps**

Letâ€™s now talk about how to perform these measurements using the PyMOL command line, which is a more efficient way to conduct multiple measurements.

### **Bond Distance Measurements**

```python
#Load the Protein
fetch 1A3N, async=0

#Hide All for a Clean Start
hide everything, all
show cartoon, all

#Measure the Distance Between Two Atoms
distance dist1, /1A3N//A/50/CA, /1A3N//A/58/CA  # Measure distance between CA atoms of residues 50 and 58 in chain A
```

### **Dihedral Angle Calculations**

```python
#Measure Dihedral Angle Between Four Atoms
dihedral dih1, /1A3N//A/50/N, /1A3N//A/50/CA, /1A3N//A/58/C, /1A3N//A/62/N
```

This measures the dihedral angle between four atoms, where dih1 is the name of the angle object. You need four atoms to define a dihedral angle.

### **Angle Measurements**

```python
#Measure Angle Between Three Atoms
angle ang1, /1A3N//A/50/CA, /1A3N//A/58/C, /1A3N//A/62/N
```

ang1 is the name for the angle object, and three atoms are used to measure the bond angle.


## **III. Python Codes**

If you want to automate this using Python scripting, this can be done using PyMOL's scripting interface, which allows you to write your entire workflow in a script.

**Python Code for Distance and Angle Measurements**

```python
import pymol
from pymol import cmd

# Step 1: Launch PyMOL
pymol.finish_launching()

# Step 2: Load the Protein Structure
cmd.fetch("1A3N")  # Load protein structure from PDB by ID

# Step 3: Hide All Elements Initially
cmd.hide("everything", "all")
cmd.show("cartoon", "all")

# Step 4: Measure Distance Between Two Atoms
cmd.distance("dist1", "/1A3N//A/50/CA", "/1A3N//A/58/CA")  # Measure distance between CA atoms of residues 50 and 58

# Step 5: Measure Dihedral Angle
cmd.dihedral("dih1", "/1A3N//A/50/N", "/1A3N//A/50/CA", "/1A3N//A/58/C", "/1A3N//A/62/N")  # Calculate dihedral angle

# Step 6: Measure Angle Between Three Atoms
cmd.angle("ang1", "/1A3N//A/50/CA", "/1A3N//A/58/C", "/1A3N//A/62/N")  # Measure angle between atoms
```

- The distance function creates an object (dist1) to measure the bond distance.
- The dihedral function calculates the dihedral angle (dih1) between four specified atoms.
- The angle function creates an object (ang1) that measures the bond angle between three atoms.

By now, you should have a good understanding of how to measure distances, angles, and dihedral angles in PyMOL using different methods. These skills are very useful when analyzing protein-ligand complexes or trying to understand protein dynamics.
