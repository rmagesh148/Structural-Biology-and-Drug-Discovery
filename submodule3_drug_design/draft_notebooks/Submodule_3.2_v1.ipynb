# **Peptide-based drug design**

As Iin the previous case of drug design, first step is to identify the biochemical pathway and which biochemical pathway is responsible for disease state. Then identify the proteins that can be targeted to modulate the biochemical pathway. If the protein is an enzyme, identify the binding site and peptide that can inhibit the binding site. In this case, you have to design a peptide that binds but does not undergo enzymatic reaction or the product is not formed (look at the section of enzyme-based drug design).  If two proteins are involved, then we need the details of interactions of two proteins.

**Scenario 1:** Protein-protein complex structure is known. In this case, we analyse the binding surface of two proteins and design a peptide drug to inhibit the interaction.

**Scenario 2:**  Protein-protein interaction is known, but the 3D structure of one or both proteins is not known.
In this case, based on the sequence of the proteins, model the protein using Alpha fold.

Use Cluspro and model the protein complex.  Identify the interacting interface of two proteins from the model.

At this stage you have to perform experiments to show that antibodies can be targeted to particular surface and those antibodies block the protein-protein interaction of interest.  The interacting surfaces are very large. Identify the peptide that can bind to antibody.

## **3.1 Example of peptide-based drug design targeting protein-protein interaction**

**Biochemical pathway**

Rheumatoid arthritis is an autoimmune disease in which our own immune system attacks cells at our joints causing inflammation and synovial membrane deformation. This results in pain, inflammation and difficulty in movement of joints. The disease is known to start with the presence of rheumatoid factor in the body. This rheumatoid factor induces production of antibodies against collagen and hence joints are attached by T cells, antibodies causing inflammation. When T cells recognize the antigen presenting cells, several adhesion molecules are responsible for generating the immune response. In the first step a protein molecule CD2 on T cells binds to CD58 on antigen presenting cells. This protein-protein interaction results in cell signaling to T cells via cytoplasmic tail of CD2 generating inflammatory cytokines and hence inflammation of joints and immune response. It is known that CD58 is highly expressed in joints in arthritis patients. If we can inhibit CD2-Cd58 interactions, we can modulate the immune response, reduce the inflammation and hence progression of arthritis.

Based on the above biochemical pathway, we can design inhibitors of CD2-CD58 protein-protein interaction that could be used as therapeutic agent for autoimmune disease rheumatoid arthritis.

In this particular case the crystal structure of the complex of CD2-CD58 is available.
