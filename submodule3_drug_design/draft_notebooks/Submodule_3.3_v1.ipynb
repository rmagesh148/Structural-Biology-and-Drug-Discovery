# **Receptor based drug design**

In drug discovery we refer to any protein that is activated by a ligand result in signal for pharmacological action as receptor. Enzymes are also receptors. However, the general definition of receptor is a biological macromolecule in or on the surface of cells to which selective binding of a specific substance results in change in conformation causing a specific physiologic effect on cells or tissues. Receptors are the largest targets of drugs in the human body including G-protein protein coupled receptors, nuclear receptors and ion channels and transporter proteins.

## **4.1 Basic concepts**
Agonists: produce a conformational change to initiate a signal that is coupled to a biological response.  Generally, they increase/enhance the signal response
Antagonists: bind to a receptor but do not produce the conformational change that initiates an intracellular signal. Generally, they decrease the receptor signal response.
Partial agonists: are able to activate a receptor but cannot produce a maximal signaling effect equivalent to that of a full agonist even when all available receptors are occupied.
Many receptors such as G-protein coupled receptors,  transport proteins and ion channel proteins are transmembrane proteins that have extracellular as well as intracellular domains.


## **4.2 Drug design targeting adrenergic receptors**

**Biochemical pathway**

Adrenergic receptors are G protein coupled receptors. They have 7-transmembrane domain structure. Among adrenergic receptors beta receptors signal heart muscle contraction, smooth muscle relaxation. Hence, these receptors are targets for drug design for controlling high blood pressure. Ligands for these receptors are adrenaline or noradrenaline. Adrenergic receptors have subtypes such as alpha and beta and further classified into alpha-1, alpha2, beta-1 and beta-2.  Without going into details of these we will focus on beta-1 adrenergic receptors that are targeted for high blood pressure drug design.

One of the difficulties with adrenergic receptors (or other G-protein coupled receptors) is that subtypes of adrenergic receptors have similar 3D structures.  We have  adrenergic receptors and subtypes, 1, 2 and subtypes of 1, 1A, 1B, 1D, 2A, 2B  and 2C. Similarly, 1, 2 and 3.
For details of structural differences read the paper Qu et al., 2019, Cell Reports 29, 29292935 December 3, 2019  2019 The Author(s). https://doi.org/10.1016/j.celrep.2019.10.088.

**Since these details cannot be covered in the current module, overall, the users/readers of this module should know that in the drug design targeting adrenergic receptors, one should consider the specificity for particular type and subtypes as binding of the designed drugs to all receptors with high or similar affinities result in side effects of the drug administered.**

For example

1. Agonists of the 2 receptors are used in the treatment of asthma (relaxation of the smooth muscles of the bronchi).
2. Antagonists of the 1 receptors are used in the treatment of hypertension and angina (slow heart and reduce force of contraction).
3. Antagonists of the 1 receptors are known to cause lowering of the blood pressure (relaxation of smooth muscle and dilation of the blood vessels).

Thus careful design of drug molecules with agonist or antagonist activity and selectivity is important.

###  **Activity: Compare Binding Sites of Alpha and Beta Adrenergic Receptors**

In this activity, you will compare the binding sites of the **alpha-adrenergic receptor** and **beta-adrenergic receptor** using their structures to identify differences that can aid in designing selective drugs.

#### **PDB IDs**:
- **Alpha-adrenergic receptor**: `9CBL`
- **Beta-adrenergic receptor**: `7BU6`

### **Steps**

#### **Step 1: Fetch the Structures**
1. Fetch both receptor structures in PyMOL:
   ```
   fetch 9CBL
   fetch 7BU6
   ```

#### **Step 2: Align the Structures**
1. Align the helical regions of `9CBL` to `7BU6`:
   ```
   9CBL > A > align > to molecule > 7BU6
   ```
2. If the molecules go out of the display window:
   Go to the top menu:
     ```
     Zoom > Visible
     ```
   Both aligned molecules will now be displayed.

#### **Step 3: Visualize Ligand Sites**
1. For `9CBL`:
   ```
   9CBL > A > ligand site > cartoon
   ```
   This zooms in on the ligand site residues of `9CBL`.

2. For `7BU6`:
   ```
   7BU6 > A > ligand site > cartoon
   ```
   This zooms in on the ligand site residues of `7BU6`.

#### **Step 4: Highlight Differences**
1. **Color the proteins**:
   Color `9CBL` (alpha-2A adrenergic receptor) in green:
     ```
     9CBL > C > green
     ```
   Color `7BU6` (beta-1 adrenergic receptor) in hot pink:
     ```
     7BU6 > C > magenta > hotpink
     ```

2. **Display specific residues**:
   In `7BU6`, select **N1363 (Asn1363)** and display it as sticks:
     ```
     select 7BU6/N1363
     show sticks
     ```
   In `9CBL`, select **F427 (Phe427)** and display it as sticks:
     ```
     select 9CBL/F427
     show sticks
     ```

#### **Step 5: Analyze the Binding Interactions**
1. **Hydrophobic residues in `9CBL`**:
   `F427 (Phe427)` is a hydrophobic residue.
   Additional hydrophobic residues near the ligand:
     ```
     F404, Y431
     ```

2. **Interactions in `7BU6`**:
   `N1363 (Asn1363)` forms a **hydrogen bond** with the ligand.
   Hydrophobic residues near the ligand:
     ```
     W1337, F1340, Y1367
     ```

**Key Insights**:
- The **hydrophobic interactions** dominate in the alpha-adrenergic receptor (`9CBL`), while the beta-adrenergic receptor (`7BU6`) has a combination of **hydrophobic** and **hydrogen bonding** interactions.
- These differences in binding sites can be exploited to design **selective drugs** for alpha- and beta-adrenergic receptors.

## **4.3 Design of drugs targeting adrenoreceptors, medication to control high blood pressure and heart diseases.**

Long ago, around 1958, Sir James proposed and used isoprenaline as -adrenoceptor-selective agonist as a template structure. A chlorine derivative dichloroisoprenaline was designed.  Isoprenaline was an agonist of -adrenoreceptor, but dichloroisoprenaline was a partial agonist of -adrenoreceptor.   By modification of the structures practolol and pronethanol, were designed as the first -blockers. However, these had severe side effects and hence a safer, more potent derivative, propranolol, was designed. Then again, selectivity was needed between 1 and 2 blockers, as 1 receptors have importance in hypertension, but 2 have importance in asthma. Practolol was designed based on the hydrogen bonding of the para position substitution at the aromatic ring.  Based on these structures and analysis of 3D structure of adrenergic receptors one can see that subtle changes in the protein 3D structure and binding surface/cavity, pharmacological activity and hence clinical application may vary.
