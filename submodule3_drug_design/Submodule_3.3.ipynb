# **Receptor based drug design**

In drug discovery we refer to any protein that is activated by a ligand result in signal for pharmacological action as receptor. Enzymes are also receptors. However, the general definition of receptor is a biological macromolecule in or on the surface of cells to which selective binding of a specific substance results in change in conformation causing a specific physiologic effect on cells or tissues. Receptors are the largest targets of drugs in the human body including G-protein protein coupled receptors, nuclear receptors and ion channels and transporter proteins.

## **4.1 Basic concepts**
Agonists: produce a conformational change to initiate a signal that is coupled to a biological response.  Generally, they increase/enhance the signal response
Antagonists: bind to a receptor but do not produce the conformational change that initiates an intracellular signal. Generally, they decrease the receptor signal response.
Partial agonists: are able to activate a receptor but cannot produce a maximal signaling effect equivalent to that of a full agonist even when all available receptors are occupied.
Many receptors such as G-protein coupled receptors,  transport proteins and ion channel proteins are transmembrane proteins that have extracellular as well as intracellular domains.


## **4.2 Drug design targeting adrenergic receptors**

**Biochemical pathway**

Adrenergic receptors are G protein coupled receptors. They have 7-transmembrane domain structure. Among adrenergic receptors beta receptors signal heart muscle contraction, smooth muscle relaxation. Hence, these receptors are targets for drug design for controlling high blood pressure. Ligands for these receptors are adrenaline or noradrenaline. Adrenergic receptors have subtypes such as alpha and beta and further classified into alpha-1, alpha2, beta-1 and beta-2.  Without going into details of these we will focus on beta-1 adrenergic receptors that are targeted for high blood pressure drug design.

One of the difficulties with adrenergic receptors (or other G-protein coupled receptors) is that subtypes of adrenergic receptors have similar 3D structures.  We have α adrenergic receptors and subtypes, α1, α2 and subtypes of α1, α1A, α1B, α1D, α2A, α2B  and α2C. Similarly, β1, β2 and β3.
For details of structural differences read the paper Qu et al., 2019, Cell Reports 29, 2929–2935 December 3, 2019 ª 2019 The Author(s). https://doi.org/10.1016/j.celrep.2019.10.088.

**Since these details cannot be covered in the current module, overall, the users/readers of this module should know that in the drug design targeting adrenergic receptors, one should consider the specificity for particular type and subtypes as binding of the designed drugs to all receptors with high or similar affinities result in side effects of the drug administered.**

For example

1. Agonists of the β2 receptors are used in the treatment of asthma (relaxation of the smooth muscles of the bronchi).
2. Antagonists of the β1 receptors are used in the treatment of hypertension and angina (slow heart and reduce force of contraction).
3. Antagonists of the α1 receptors are known to cause lowering of the blood pressure (relaxation of smooth muscle and dilation of the blood vessels).

Thus careful design of drug molecules with agonist or antagonist activity and selectivity is important.

Structure of β-adrenergic receptor with natural ligand norepinephrine shows that the ligand forms four hydrogen bonds with the receptor (Ser, Asn and Asp side chains) and hydrophobic interaction of catechol ring with Phe and Trp side chains as shown in the structure here.


## **PML Script: Visualize Beta Adrenergic Receptor with Ligand Norepinephrine (7BU6)**

````python
with open("3.3 Beta1_Norepinephrine_Binding.pml", "w") as scriptout:

    # Step 1: Fetch the Structure
    scriptout.write("fetch 7BU6, receptor_ligand, async=0\n")  # Fetch the structure of beta1 adrenergic receptor
    scriptout.write("hide everything, all\n")  # Hide all default representations
    scriptout.write("show cartoon, receptor_ligand\n")  # Show the receptor as a cartoon
    scriptout.write("color cyan, receptor_ligand\n")  # Color the receptor cyan
    scriptout.write("zoom receptor_ligand\n")  # Zoom into the receptor

    # Step 2: Isolate the Protein and Ligand
    scriptout.write("remove solvent\n")  # Remove solvent molecules
    scriptout.write("remove inorganic\n")  # Remove inorganic ions
    scriptout.write("select norepinephrine, resn E5E\n")  # Select the ligand labeled as E5E
    scriptout.write("create nor, norepinephrine\n")  # Save the ligand as an object named 'nor'
    scriptout.write("show sticks, nor\n")  # Show the ligand in stick representation
    scriptout.write("color yellow, nor\n")  # Color the ligand yellow
    scriptout.write("zoom nor\n")  # Focus on the ligand

    # Step 3: Highlight Nearby Residues
    scriptout.write("select nearby_residues, byres all within 5 of nor\n")  # Select residues within 5 Å of the ligand
    scriptout.write("show sticks, nearby_residues\n")  # Show nearby residues in stick representation
    scriptout.write("color magenta, nearby_residues\n")  # Color nearby residues magenta

    # Step 4: Create the Active Site
    scriptout.write("create active_site, nearby_residues\n")  # Save nearby residues as 'active_site'
    scriptout.write("show surface, active_site\n")  # Display the active site as a surface
    scriptout.write("set transparency, 0.3, active_site\n")  # Set transparency for clarity
    scriptout.write("color red, active_site\n")  # Color the active site surface red
    scriptout.write("zoom (nor or active_site)\n")  # Focus on the ligand and the active site

    # Step 5: Save the Visualization Outputs
    scriptout.write("png beta1_binding_visualization.png, dpi=300\n")  # Save a high-resolution PNG of the visualization
    scriptout.write("save beta1_binding_session.pse\n")  # Save the PyMOL session for future reference

    # Notes for the user
    scriptout.write("# This script visualizes the beta1 adrenergic receptor (7BU6) with norepinephrine (E5E).\n")
    scriptout.write("# Steps include ligand isolation, nearby residue highlighting, and active site visualization.\n")
````

**Key Insights**:

## **PML Script: Compare Alpha (9CBL) and Beta (7BU6) Adrenergic Receptors**

````python
with open("3.3 Adrenergic_Receptor_Comparison.pml", "w") as scriptout:

    # Step 1: Fetch the Structures
    scriptout.write("fetch 9CBL, alpha_receptor, async=0\n")  # Fetch alpha-adrenergic receptor structure
    scriptout.write("fetch 7BU6, beta_receptor, async=0\n")  # Fetch beta-adrenergic receptor structure
    scriptout.write("hide everything, all\n")  # Hide all default representations
    scriptout.write("show cartoon, all\n")  # Show both receptors in cartoon representation
    scriptout.write("color green, alpha_receptor\n")  # Color alpha receptor green
    scriptout.write("color hotpink, beta_receptor\n")  # Color beta receptor hot pink
    scriptout.write("zoom all\n")  # Zoom to fit both receptors

    # Step 2: Align the Structures
    scriptout.write("align alpha_receptor and name CA, beta_receptor and name CA\n")  # Align alpha to beta receptor
    scriptout.write("zoom all\n")  # Adjust view to see both structures

    # Step 3: Visualize Ligand Sites
    scriptout.write("select alpha_ligand_site, alpha_receptor and (byres all within 5 of resn LIG)\n")  # Select ligand site of alpha receptor
    scriptout.write("select beta_ligand_site, beta_receptor and (byres all within 5 of resn E5E)\n")  # Select ligand site of beta receptor
    scriptout.write("show sticks, alpha_ligand_site\n")  # Show ligand site residues as sticks for alpha receptor
    scriptout.write("show sticks, beta_ligand_site\n")  # Show ligand site residues as sticks for beta receptor
    scriptout.write("color yellow, alpha_ligand_site\n")  # Color alpha ligand site residues yellow
    scriptout.write("color cyan, beta_ligand_site\n")  # Color beta ligand site residues cyan
    scriptout.write("zoom alpha_ligand_site or beta_ligand_site\n")  # Focus on ligand sites

    # Step 4: Highlight Specific Residues
    scriptout.write("select alpha_key_residues, alpha_receptor and resi 404+427+431\n")  # Hydrophobic residues in alpha receptor
    scriptout.write("select beta_key_residues, beta_receptor and resi 1337+1340+1363+1367\n")  # Key residues in beta receptor
    scriptout.write("show sticks, alpha_key_residues\n")  # Show key residues in alpha receptor as sticks
    scriptout.write("show sticks, beta_key_residues\n")  # Show key residues in beta receptor as sticks
    scriptout.write("color orange, alpha_key_residues\n")  # Color key residues in alpha receptor orange
    scriptout.write("color magenta, beta_key_residues\n")  # Color key residues in beta receptor magenta

    # Step 5: Display Specific Residues
    scriptout.write("select alpha_phe427, alpha_receptor and resi 427\n")  # Select Phe427 in alpha receptor
    scriptout.write("select beta_asn1363, beta_receptor and resi 1363\n")  # Select Asn1363 in beta receptor
    scriptout.write("show sticks, alpha_phe427\n")  # Show Phe427 as sticks
    scriptout.write("show sticks, beta_asn1363\n")  # Show Asn1363 as sticks
    scriptout.write("color red, alpha_phe427\n")  # Color Phe427 red
    scriptout.write("color blue, beta_asn1363\n")  # Color Asn1363 blue

    # Step 6: Save Visualization Outputs
    scriptout.write("zoom all\n")  # Adjust zoom to fit everything
    scriptout.write("png adrenergic_receptor_comparison.png, dpi=300\n")  # Save high-resolution PNG
    scriptout.write("save adrenergic_receptor_comparison_session.pse\n")  # Save PyMOL session for future use

    # Notes for the user
    scriptout.write("# This script compares alpha (9CBL) and beta (7BU6) adrenergic receptor ligand binding sites.\n")
    scriptout.write("# Hydrophobic interactions dominate in 9CBL, while hydrogen bonding and hydrophobic interactions are present in 7BU6.\n")
````

## **4.3 Design of drugs targeting adrenoreceptors, medication to control high blood pressure and heart diseases.**

Long ago, around 1958, Sir James proposed and used isoprenaline as β-adrenoceptor-selective agonist as a template structure. A chlorine derivative dichloroisoprenaline was designed.  Isoprenaline was an agonist of β-adrenoreceptor, but dichloroisoprenaline was a partial agonist of β-adrenoreceptor.   By modification of the structures practolol and pronethanol, were designed as the first β-blockers. However, these had severe side effects and hence a safer, more potent derivative, propranolol, was designed. Then again, selectivity was needed between β1 and β2 blockers, as β1 receptors have importance in hypertension, but β2 have importance in asthma. Practolol was designed based on the hydrogen bonding of the para position substitution at the aromatic ring.  Based on these structures and analysis of 3D structure of adrenergic receptors one can see that subtle changes in the protein 3D structure and binding surface/cavity, pharmacological activity and hence clinical application may vary.
